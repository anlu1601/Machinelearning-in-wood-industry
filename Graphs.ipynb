```python
import timeit
import matplotlib.pyplot as plt
import os
import time
import numpy as np
import cv2
```

```python

X = ['0.5', '0.6', '0.7', '0.8', '0.9']
scale_factor = 0.8

sgd = [0.97512, 0.97674, 0.96204, 0, 0]
adam = [0.93091, 0.89321, 0.81905, 0.87139, 0.94238]
sgd = [s - scale_factor if s - scale_factor > 0 else 0 for s in sgd]
adam = [s - scale_factor if s - scale_factor > 0 else 0 for s in adam]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, sgd, 0.4, scale_factor, label = 'SGD')
plt.bar(X_axis + 0.2, adam, 0.4, scale_factor, label = 'Adam')

plt.annotate(sgd[1]+scale_factor, (1, sgd[1]+scale_factor))
  
plt.xticks(X_axis, X)
plt.xlabel("Dropout")
plt.ylabel("F1 Score")
plt.title("F1 Score - 50 Epochs - 80%+")
plt.legend()
plt.show()
```
